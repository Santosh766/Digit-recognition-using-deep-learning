```python
x_train  = x_train/255   # scaling all the pixel data for better prediction
x_test  = x_test/255
```

```python
x_train_flattened = x_train.reshape(len(x_train), 28*28)
#x_test_flattened = x_test.reshape(len(x_test), 28*28)
```

```python
x_test_flattened = x_test.reshape(len(x_test),28*28 )
```

Very simple neural network with no hidden layers

```python
model  = keras.Sequential([
    keras.layers.Dense(10, input_shape= (784,), activation = "sigmoid")
])

model.compile(optimizer = 'adam', 
            loss = "sparse_categorical_crossentropy",
             metrics = ['accuracy'])

model.fit(x_train_flattened, y_train, epochs=5)
```

In the above model we can see that log loss is decreasing whereas the accuracy is increasing that basically show that model is trained for the train dataset and uses its experience to predict with less error. Here we have not used any hidden layer just input and output layer is used.

```python
model.evaluate(x_test_flattened, y_test)
```

# With hidden layer 

```python
model  = keras.Sequential([
    keras.layers.Dense(200, input_shape=(784,), activation='relu'),   # 100 hidden layers are imposed with ReLU activation
    keras.layers.Dense(10, input_shape= (784,), activation = "sigmoid")
])

model.compile(optimizer = 'adam', 
            loss = "sparse_categorical_crossentropy",
             metrics = ['accuracy'])

model.fit(x_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(x_test_flattened, y_test)
```

```python
plt.matshow(x_test[0])
```

```python
y_pred = model.predict(x_test_flattened)
y_pred[:5]
```

The index with maximum value at the output layer will be treated as the final output

```python
np.argmax(y_pred[0])
```

```python
y_test[:5]
```

```python
y_pred[:5]
```

```python
y_pred_labels =  [np.argmax(i) for i in y_pred]
y_pred_labels[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred_labels)
plt.figure(figsize = (10,8))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
